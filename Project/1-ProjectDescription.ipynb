# Project Description — MME 9710: Advanced Computational Fluid Dynamics

## 1. Project Overview

This project provides hands-on experience with the OpenFOAM CFD framework (using [openfoam.org](https://openfoam.org)). Each student will be assigned an incompressible flow tutorial case and is required to:

1. Reproduce the baseline result and compare it to literature
2. Perform a grid-independence study
3. Conduct a small parametric study by systematically varying one parameter of the simulation

The project emphasizes **physical understanding**. Students are expected to explain observed trends in results using theory covered in the course. Simply producing plots is not sufficient - every result must be interpreted in the context of the governing equations and flow physics.

<div class="alert alert-warning">

**Warning:** Physical interpretation is the most heavily weighted component of this project. Results that are not explained using course theory will receive partial credit at best, even if the simulations were run correctly.

</div>

## 3. Project Workflow

### Project Proposal

Before beginning the parametric study, each student must submit a brief written proposal for instructor approval. The purpose of the proposal is to ensure the chosen parametric study parameter is physically meaningful, that the planned parameter range is appropriate, and that the student has a clear plan before investing time in simulations.

The proposal should be **no longer than a single page, double-spaced**, and must address:

- Parametric parameter choice
- Proposed parameter range and justification (prediction of impact)
- Validation approach
- Grid independence parameter

<div class="alert alert-success">

**Tip:** Proposals will be reviewed and returned with feedback within one week.

</div>

### Choosing Your Parameter(s)

Students must select a single parameter to vary in a systematic and physically meaningful way. The choice of parameter must be **justified on physical grounds** — it should be something that meaningfully changes the flow physics and for which you can form a prediction before running (note that the prediction does not have to be correct).

Acceptable parameters to vary include:

- **Boundary conditions:** inlet velocity, lid velocity, or pressure gradient (effectively changing the Reynolds number)
- **Geometric parameters:** domain aspect ratio, step height (if applicable)
- **Numerical settings:** time stepping strategy, discretization schemes in `fvSchemes`, etc.

## 4. Project Report

Each student submits an individual written report. The report should be concise, well-structured, and written for an audience of CFD engineers.

**Suggested length:** 8 to 10 double-spaced, single-sided pages including figures. Quality of analysis is valued over length.

The following sections are required (note you may change the name of the sections to suit your preference):

---

### 4.1 Introduction

Briefly introduce your topic and any relevant literature or information that you will be using for the parametric study. This does not have to be exhaustive, but should provide the reader with a general overview of the topic.

---

### 4.2 Numerical Setup

Describe the geometry, mesh, boundary conditions, solver settings, and time-stepping methods used in your simulations.

---

### 4.3 Mesh Sensitivity Study

A mesh sensitivity study must be completed **before** any parametric variation. This is a fundamental step in any credible CFD workflow and ensures that your results are free from discretization error.

Using the `blockMesh` file within each case, complete the following:

1. Run at least three mesh resolutions using the **Celik et al.** approach discussed in class. Note that increases in mesh density do not have to double the previous.
2. Report a key quantity as a function of mesh size
3. Identify the mesh resolution at which results are sufficiently grid-independent and justify this choice
4. Use this mesh for all subsequent parametric runs

<div class="alert alert-warning">

**Warning:** All subsequent parts of the project must be run on the mesh identified as sufficiently grid-independent in this step.

</div>

---

### 4.4 Parametric Study

Present your parametric study results clearly and systematically. Every figure must have axis labels, units, and a descriptive caption.

---

### 4.5 Discussion

Provide a physical interpretation of trends observed in your results. Compare your findings with your pre-study predictions and reference data. If solver settings were varied, discuss their impact here.

---

### 4.6 References

Cite all sources used in your report.

## 6. Important Notes

### On the Parametric Study

A good parametric study is not just a collection of runs — it is a **structured investigation of how and why a flow changes**. The parameter range must be chosen so that observable, explainable trends emerge. For example:

- A Reynolds number sweep should span regimes where the flow structure visibly evolves (e.g., from a symmetric single vortex to an asymmetric multi-vortex system in the cavity case)
- A geometric variation should be large enough that the influence on the flow pattern is clear, not just a small shift in velocity magnitude

<div class="alert alert-warning">

**Warning:** If your trend plot is a flat line or shows no clear behaviour, your parameter range is likely too narrow. Consult the instructor before finalizing your parameter choices.

</div>

### On Physical Reasoning

Every result you present must be accompanied by an explanation rooted in the governing equations or flow theory. Ask yourself: *given what I know about this flow, does this result make sense?*

If Re increases, what do you expect to happen to the boundary layer thickness, the pressure gradient, the vortex structure? These are the questions your discussion must answer.

<div class="alert alert-warning">

**Warning:** Statements such as *"as velocity increases, the flow changes"* are not acceptable. Specific, quantitative, and theory-backed reasoning is expected at the graduate level.

</div>
