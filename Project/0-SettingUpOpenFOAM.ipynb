# Setting Up OpenFOAM

OpenFOAM (Open Field Operation and Manipulation) is an open-source CFD toolbox widely used in both industry and academia. This notebook provides installation guidance for all major operating systems.

There are two main distributions of OpenFOAM:

| Distribution | Maintained by | Website |
|:---:|:---:|:---:|
| **OpenFOAM** | The OpenFOAM Foundation | [openfoam.org](https://openfoam.org) |
| **OpenFOAM** | ESI Group (OpenCFD) | [openfoam.com](https://www.openfoam.com) |

Both distributions are free and open-source. For this course we will use the **OpenFOAM Foundation** version ([openfoam.org](https://openfoam.org)), which is the original distribution and most commonly referenced in academic literature.

In terms of which version to use, I will be using OF 12. Please feel free to use OF 13, however, note that there might be slight variations in terms of file paths, solver settings etc.

<div class="alert alert-warning">

**Warning:** OpenFOAM runs natively on **Linux** only. macOS and Windows users will need to follow additional steps outlined below.

</div>

## Linux

Linux is the recommended platform for OpenFOAM and has the most straightforward installation process. Ubuntu is the most commonly used distribution and has the best support.

### Ubuntu / Debian

The OpenFOAM Foundation provides an official Ubuntu package repository. Follow the official instructions here:

**[https://openfoam.org/download/ubuntu](https://openfoam.org/download/ubuntu)**

In brief, the installation involves:
1. Adding the OpenFOAM repository to your system
2. Running `sudo apt install openfoam12` (replace `12` with the current version.)
3. Adding the OpenFOAM environment to your shell profile

### Other Linux Distributions

For other distributions (Fedora, CentOS, Arch, etc.), the recommended approach is to compile from source:

**[https://openfoam.org/download/source](https://openfoam.org/download/source)**

<div class="alert alert-success">

**Tip:** If you are new to Linux, Ubuntu 22.04 LTS is the recommended starting point — it has the widest package support and the most online resources.

</div>

## Verifying Your Installation

Once OpenFOAM is installed, verify it is working correctly by running the built-in cavity tutorial:

```bash
# Source the OpenFOAM environment (add this to your .bashrc to make it permanent)
source /opt/openfoam12/etc/bashrc

# If you have already done so, make sure your Run directory is set up:
mkdir -p $FOAM_RUN

# Copy the cavity tutorial to your working directory
cp -r $FOAM_TUTORIALS/incompressibleFluid/cavity .
cd cavity

# Run the case
foamRun
```

If the simulation runs without errors, your installation is working correctly.
